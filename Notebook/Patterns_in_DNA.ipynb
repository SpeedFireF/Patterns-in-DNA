```python
import random
import matplotlib.pyplot as plt

palindrome_locations = []
with open('/Users/damirabdulaev/optional_assignment/hcmv.data.txt', 'r') as f:
    for line in f:
        palindrome_locations.append(int(line.strip()))

# Define DNA sequence length and number of palindromes to simulate
dna_length = 229354
n_palindromes = 296

# Simulate random scatter of palindromes
arrays = []
for i in range(4):
    random.seed(i + 1)
    arr = random.sample(range(dna_length), n_palindromes)
    arrays.append(arr)

# Plot real and simulated palindrome locations
fig, ax = plt.subplots()
ax.vlines(x=palindrome_locations, ymin=0.8, ymax=1, color='red', linewidth=0.5)
for i, arr in enumerate(arrays):
    if i != 3:
        ax.vlines(x=arr, ymin=0.2 + i * 0.2, ymax=0.4 + i * 0.2, color='blue', linewidth=0.5)
ax.set_xlabel('DNA position')
ax.set_yticks([0.3, 0.5, 0.7, 0.9])
ax.set_yticklabels(['Simulated 3', 'Simulated 2', 'Simulated 1', 'Real'])
plt.show()
```

### Random scatter results:
#### Based on scatter plot we can see that real palindromes are not randomly scattered, because they are not evenly distributed across the DNA sequence. Although we can see some gaps, but they are not even. For example, we can see that there are more palindromes in the first 100000 positions than in the last 100000 positions.

```python
# Compute spacings between consecutive palindromes
arrays[0] = sorted(arrays[0])
for j in range(1, 10):
    palindrome_spacings_random = [arrays[0][i + j] - arrays[0][i + j - 1] for i in range(0, len(arrays[0]) - j - 1, j)]

    palindrome_spacings = [palindrome_locations[i + j] - palindrome_locations[i + j - 1] for i in
                           range(0, len(palindrome_locations) - j - 1, j)]

    pair_spacings_random = [sum(palindrome_spacings_random[i:i + 2]) for i in
                            range(len(palindrome_spacings_random) - 1)]
    # Compute sums of consecutive pairs of palindrome spacings
    pair_spacings = [sum(palindrome_spacings[i:i + 2]) for i in range(len(palindrome_spacings) - 1)]

    fig, ax = plt.subplots()
    ax.hist(palindrome_spacings, bins=range(2, 5000, 100))
    ax.hist(palindrome_spacings_random, bins=range(2, 5000, 100))

    ax.set_xlabel('Sum of consecutive pairs of palindrome spacings')
    ax.set_ylabel('Count')
    plt.show()
```

#### Based on the histogram we can see that spacing in CMV DNA is different with the random scatter. While random scatter seems to be more uniform, the real scatter is more concentrated around 1000. This means that the real scatter is more likely to have palindromes that are close to each other.
